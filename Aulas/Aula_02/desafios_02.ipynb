## Inicialização do Dataframe

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/pedrogrigorio/Data-Immersion-Alura/main/Materiais/dados_imoveis.csv"
dados = pd.read_csv(url)
dados
```

```python
dados[["Moeda", "Valor_anuncio", "Tipo_anuncio"]] = dados["Valor"].str.split(expand = True)
dados_vendas = dados[dados["Tipo_anuncio"].isnull()]
dados_vendas["Valor_anuncio"] = dados_vendas["Valor_anuncio"].str.replace(".","").astype(float)
dados_vendas
```

```python
dados_vendas.info()
```

## Desafio 01

### Criar uma escala de valores em milhões.

Para isso, basta dividir os valores da coluna Valor_anuncio por 1 milhão e, em seguida, organizar pelo menor valor para saber qual precisão de casas decimais usar.

```python
dados_vendas["Valor_anuncio"].sort_values().div(1000000).round(3)
```

```python
dados_vendas["Valor_em_milhão"] = dados_vendas["Valor_anuncio"].sort_values().div(1000000).round(3)
dados_vendas
```

## Desafio 02

### Deixar o gráfico do histograma de valores legível (alterar labels, cores, título, escala).

Para a escala, podemos usar a criada no desafio anterior e a cor resolvi colocar verde escuro, para contrastar bem com o fundo claro. Além disso, a alteração de labels e título é feita com os comandos `set_title`, `set_xlabel` e `set_ylabel`.

```python
sns.set()
plt.figure(figsize=(10,8))
ax = sns.histplot(data = dados_vendas, x = "Valor_em_milhão", color='darkgreen')
ax.set_title("Valores de Imóveis em São Paulo", size = 16)
ax.set_xlabel("Valor em Milhões", size = 14)
ax.set_ylabel("Quantidade de Imóveis", size = 14)
plt.xlim((-0.5, 20))
plt.show()
```

## Desafio 03

### Preço do metro quadrado por bairro e plotar em um gráfico ideal.

Basta criar uma outra coluna que receba o resultado da divisão da coluna Valor_anuncio pela coluna Metragem e depois tirar uma média por bairro.

```python
dados_vendas["Valor_metro"] = dados_vendas["Valor_anuncio"] / dados_vendas["Metragem"]
dados_vendas.head()
```

```python
preco_metro_bairro = dados_vendas.groupby("Bairro")["Valor_metro"].mean()
preco_metro_bairro
```

Criando um dataframe a partir da series.

```python
df_preco_metro_bairro = pd.DataFrame(preco_metro_bairro, columns=['Valor_metro'])
df_preco_metro_bairro.sort_values("Valor_metro").round(2)
```

Criando um dataframe para os 25 primeiros e os 25 últimos do dataframe acima.

```python
df_25maiores = df_preco_metro_bairro.head(25).sort_values("Valor_metro",ascending = True).round(2)
df_25menores = df_preco_metro_bairro.tail(25).sort_values("Valor_metro",ascending = True).round(2)
```

Buscando algumas informações sobre os dataframes que acabamos de criar.

```python
df_25maiores.describe()
```

```python
df_25menores.describe()
```

Plotando os gráficos.

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(y=df_25maiores.index, x="Valor_metro", data=df_25maiores, orient="h", palette="Blues_d")
ax.set_title("25 Maiores Valores de Metro por Bairro")
ax.set_xlabel("Valor do Metro")
ax.set_ylabel("Bairro")
plt.xlim((0, 13000))
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(y=df_25menores.index, x="Valor_metro", data=df_25menores, orient="h", palette = "Blues_d")
ax.set_title("25 Menores Valores de Metro por Bairro")
ax.set_xlabel("Valor do Metro")
ax.set_ylabel("Bairro")
plt.xlim((0, 13000))
plt.show()
```

## Desafio 05

### Pesquisar um visualização para analisar os quartis, mediana e outliers.

Um bom gráfico para verificar essas informações é o de caixa. Nesse sentido, iremos primeiro buscar algumas informações para depois verificar no gráfico.

```python
dados_vendas["Valor_metro"].describe()
```

O método `DataFrame.describre()` nos fornece os quartis, mediana e os outliers, agora, podemos plotar o gráfico e verificar essas informações.

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x = "Valor_metro", data=dados_vendas, linewidth=2.5, palette="Blues_d")
ax.set_title("Valor do m² dos imóveis de São Paulo")
ax.set_xlabel("Valor do Metro")
plt.show()
```

Olhando par ao gráfico acima, percebemos que existe uma grande quantidade de outliers e que a maior parte dos dados estão concentrados entre 0 e 10000. Nesse sentido, irei plotar outro gráfico, limitando o eixo x, para verificar melhor esses dados predominantes.

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x = "Valor_metro", data=dados_vendas, linewidth=2.5, palette="Blues_d")
ax.set_title("Valor do m² dos imóveis de São Paulo")
ax.set_xlabel("Valor do Metro")
plt.xlim((0, 16600))
plt.show()
```

No gráfico acima, conseguimos notar que o primeiro quartil está entre 4 e 6 mil, a mediana está abaixo de 7 mil e o terceiro quartil pouco abaixo de 10 mil.
